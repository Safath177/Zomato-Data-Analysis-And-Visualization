```python
import pandas as pd
import plotly.graph_objects as px
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
zomato_df = pd.read_csv('/content/zomato.csv')
zomato_df
```

```python
zomato_c_c = pd.read_csv('/content/zomato_country_data.csv')
zomato_c_c
```

# Merging two datasets


```python
zomato = pd.merge(zomato_df, zomato_c_c, on='Country Code')
zomato
```

```python
zomato.info(
)
```

# Converting cost to INR

```python
zomato['Converted_Cost'] = zomato['Average Cost for two'] * zomato['Exchange Rate']
zomato.head()
```

# Splitting the cuisine columns

```python
zomato['Cuisines'] = zomato['Cuisines'].astype(str)
```

```python
zomato['Cuisines'] = zomato['Cuisines'].str.split(', ')
```

```python
zomato_exp = zomato.explode('Cuisines')
zomato_exp.head()
```

```python
cuisine_count = zomato_exp['Cuisines'].value_counts()
cuisine_count
```

```python
zomato_exp.info()
```

```python
zomato_exp.to_csv('zomato_final.csv',index = False)
```

# EDA

```python
df = pd.read_csv('/content/zomato_final.csv')
df.head()
df.info()
```

```python
df.describe()
```

```python
# distribution of ratings

sns.histplot(df['Aggregate rating'], bins =20, kde = True)
plt.title('Distribution of Ratings')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.show()
```

```python
# count of restaurants in each country

sns.countplot(df['Country'])
plt.title('Count of Restaurants in Each Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

```python
# scatter plot between rating and votes

sns.scatterplot(x = 'Aggregate rating', y = 'Votes', data = df)
plt.title('Rating vs Votes')
plt.xlabel('Rating')
plt.ylabel('Votes')
plt.show()
```

```python
# correlation heatmap

col = ['Aggregate rating','Votes','Average Cost for two','Converted_Cost','Price range']
corr = df[col].corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# restaurants with online delivery

sns.countplot(df['Has Online delivery'])
plt.title('Restaurants with Online Delivery')
plt.xlabel('Has Online Delivery')
plt.ylabel('Count')
plt.show()
```

```python
# comparing ratings and price range

sns.boxplot(x = 'Price range', y = 'Aggregate rating', data = df)
plt.title('Ratings vs Price Range')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.show()
```
